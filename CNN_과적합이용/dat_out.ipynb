```python
# 만든사람 : 정성모
# 원 데이터 : data.out, 독립변수 : data.out(파티션ID, Idistance Key, 참조점으로부터 거리), 종속변수 : Random
# 데이터가 정형화 되어 있지않아 전처리 과정후, Conv1D layer를 이용하여 데이터를 학습시켜 데이터 분류 실험'n# 네트워크에 파라미터 개수 증가, Dense layer 순차적으로 filter 쌓기, ResNet 실험

import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt


%matplotlib inline

# 데이터 전처리 과정으로 데이터 중간에 정형화 되어 있지 않은 데이터 row 삭제
f = open("C:/Users/jsm97/Desktop/data/data.out")
X=[]
while True:
    line = f.readline()
    if not line:
        break
    temp = line.split(",")
    if len(temp) == 5:
        X.append(temp)

pdData = pd.DataFrame(X)
pdData1 = pdData.drop([2,4], axis =1)

pdData1[0] = pd.to_numeric(pdData1[0], errors='coerce')
pdData1[1] = pd.to_numeric(pdData1[1], errors='coerce')
pdData1[3] = pd.to_numeric(pdData1[3], errors='coerce')

pdData1.dropna(inplace=True)

y = np.random.randint(0, 10, size=(9974,))

x = pdData1.to_numpy()
x = x[x[:,2] < 1500]
#plt.scatter(x[:,2], x[:,0], s=1)
x_max = np.max(x, axis=0)
x_max[0] = 1

x = x / x_max

x = x.reshape(9974,3,1,1)
print(x)
print(y)


'''
y = pdData1.drop([1,3], axis=1)
y = y.to_numpy()
y = y.reshape(9975,)

x = pdData1.drop([0], axis=1)
x = x.to_numpy()
print(x)
x_max = np.max(x, axis = 0)
x = x / x_max
print(x)
x = x.reshape(9975,2,1)
print(x.shape)
'''
```

```python
# 파라미터 개수 증가 및 Dense layer 순차적으로 layer 쌓은 모델 test
from tensorflow.keras import models, layers, utils

print(y.size)
model = models.Sequential()
model.add(layers.Conv1D(32, 1, activation='relu',input_shape=(3,1)))
model.add(layers.Conv1D(64, 1, activation='relu'))
model.add(layers.Conv1D(128, 1, activation='relu'))
model.add(layers.Conv1D(256, 1, activation='relu'))
model.add(layers.Conv1D(512, 1, activation='relu'))
model.add(layers.Conv1D(1024, 1, activation='relu'))
model.add(layers.Conv1D(2048, 1, activation='relu'))
model.add(layers.Conv1D(4096, 1, activation='relu'))
model.add(layers.Conv1D(8192, 1, activation='relu'))

model.add(layers.Flatten())
model.add(layers.Dense(2048, activation='relu'))
model.add(layers.Dense(1024, activation='relu'))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))

model.add(layers.Dense(10, activation='softmax'))

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.summary()

#odel.fit(x, y, epochs=20, batch_size=None)
```
